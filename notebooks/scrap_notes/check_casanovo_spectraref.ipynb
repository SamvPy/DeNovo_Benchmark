https://github.com/Noble-Lab/casanovo/issues/309

```python
from denovo_utils.parsers import DenovoEngineConverter
```

```python
filename = "S14_Rep2"

path_denovo_output = f"/home/samva/Doctorate/data_directory/denovo_project/denovo_results/casanovo/{filename}.mztab"
path_mgf = f"/home/samva/Doctorate/data_directory/denovo_project/mgf_filtered/{filename}.mgf"
```

```python
parser = DenovoEngineConverter.select("casanovo")
result = parser.parse(
    result_path=path_denovo_output,
    mgf_path=path_mgf,
    max_length=30
).to_dataframe()
```

```python
indices = result["metadata"].apply(lambda x: int(x["spectra_ref"].split("=")[1])).tolist()

previous_i = -1
skipped_i = []
for i in indices:
    if i != previous_i+1:
        skipped_i.append(i)
    previous_i = i
```
