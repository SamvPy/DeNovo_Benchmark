```python
result_root = "/home/samva/Doctorate/data_directory/denovo_project/denovo_results"
mgf_root = "/home/samva/Doctorate/data_directory/denovo_project/mgf_filtered"
filename = "F01_Fraction1"

search_engines = [
    ("casanovo", ".mztab"),
    ("contranovo", ".mztab"),
    ("instanovo", ".csv"),
    ("novob", ".tsv"),
    ("pepnet", ".tsv"),
    # "novor",
    # "pepnovo",
    # "directag"
]

psmlists = []
mgf_file = mgf.read(os.path.join(mgf_root, filename + ".mgf"))

for search_engine in search_engines:
    print(search_engine)
    result_path = os.path.join(result_root, search_engine[0], filename + search_engine[1])

    parser = DenovoEngineConverter.select(search_engine[0])
    psmlist = parser.parse(
        result_path=result_path,
        mgf_path=os.path.join(mgf_root, filename+".mgf")
    )
    psmlists.append(psmlist.to_dataframe())

psms_df = pd.concat(psmlists, ignore_index=True)
psms_df["peptidoform_str"] = psms_df.peptidoform.apply(lambda x: x.proforma)
```
